```python
## WHERE CONDITION
## where(logic, T, F)

x = -10
print(np.where(x>0 , "positive", "neg"))
```

```python
stores['stores_level'] = np.where(stores['Total number of estb']<10000, 'Low',
                                np.where((stores['Total number of estb']>=10000) & (stores['Total number of estb']<30000),
                                'Medium','High'))
```

```python
stores.head(20)
```

```python
## APPLYING CONSTITIONS ON A DATAFRAM USING LOC[,]

stores.loc[stores.StoreType == "Super Market" , ["StoreType","OperatingCost"]]
```

```python
stores.loc[(stores.OperatingCost > 20), ["StoreType", "OperatingCost"]]
```

```python
stores.loc[((stores.OperatingCost>20) & (stores.OperatingCost<30)), ["StoreType", "OperatingCost"]]
```

```python
stores.loc[(stores.StoreType == "Super Market") | (stores.StoreType == "Electronics") , ["StoreType","OperatingCost"]]
```

```python
stores.loc[(stores.OperatingCost<30) & (stores.StoreType == "Super Market"), ["StoreType","OperatingCost"]]
```
