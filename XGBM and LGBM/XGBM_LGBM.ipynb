```python
# First XGBoost model for Pima Indians dataset
from numpy import loadtxt
from pandas import read_csv
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [25, 12]
plt.rcParams['axes.grid'] = True
plt.gray()
```

```python
# Load data
# dataset = loadtxt('/content/pima-indians-diabetes.data.csv', delimiter=",")
col_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataset = read_csv(r"C:\Users\Bhaskar Kumbhar\ExcelR\Artificial Intelligence Course\XGBM and LGBM\pima-indians-diabetes.data.csv")
array = dataset.values
# split data into X and y
X = array[:,0:8]
Y = array[:,8]
```

```python
dataset
#IN class 1 - Diabetes detectd and 0 - Diabetes not detected
```

```python
dataset.shape
```

```python
# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=7)
```

```python
# fit model no training data
model = XGBClassifier(n_estimators=100, max_depth=3)
model.fit(X_train, y_train)
```

```python
# Predict on training data
y_pred = model.predict(X_train)
accuracy = accuracy_score(y_train, y_pred)
print(accuracy)
```

```python
# Make predictions for test data
y_pred = model.predict(X_test)
# Evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn import metrics
params = {
    'n_estimators': [200, 250, 400, 450, 500, 600],
    'learning_rate': [0.0001, 0.001, 0.01, 0.1],
    'max_depth': range(3, 14),
    'colsample_bytree': [0.4, 0.5, 0.6, 0.8]
}

xgb_estimator = RandomizedSearchCV(XGBClassifier(), param_distributions=params, cv=5, random_state=0, n_jobs=-1)
xgb_estimator.fit(X_train, y_train)

y_pred1 = xgb_estimator.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred1)
print("test accuracy : ", accuracy1)

y_pred2 = xgb_estimator.predict(X_train)
accuracy2 = accuracy_score(y_train, y_pred2)
print("train accuracy : ", accuracy2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# split data into train and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(x_train, label= y_train)
```

```python
params = {}
params['learning_rate'] = 0.002
params['boosting_type'] = 'gbdt' # 'dart'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10
```

```python
clf = lgb.train(params, d_train, 1400)
```

```python
# Prediction
y_pred=clf.predict(x_test)
```

```python
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
```
